```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
k_values = list(range(1, 11))
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k for KNN Classifier')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

# Assuming you have your dataset 'X' and labels 'y'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
k_values = list(range(1, 11))
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    # Compute the confusion matrix for the test data
    cm_test = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix (Test Data) for k={k}:\n", cm_test)

plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k for KNN Classifier')
plt.grid(True)
plt.show()

# Optionally, you can also compute the confusion matrix for the training data
y_train_pred = knn.predict(X_train)
cm_train = confusion_matrix(y_train, y_train_pred)
print("Confusion Matrix (Training Data):\n", cm_train)
```
